```python
from ultralytics import YOLO

if __name__ == "__main__":
    # Path ke file dataset YAML
    dataset_yaml = "/kaggle/input/trashclassyolo/data.yaml"

    # Load Pre-trained YOLOv11m Model
    model = YOLO("yolo11m.pt")

    # Training
    model.train(
        data=dataset_yaml,
        epochs=50,
        batch=16,
        imgsz=640,
        workers=4, 
        name="trash_detection_yolo",
        device="cuda",
        augment=True,
    )

    # Evaluasi
    results = model.val()
    print("Validation Results:", results)

    # Ekspor Model (Opsional)
    model.export(format="onnx")  # Ekspor ke format ONNX
```

```python
dataset_yaml = "/kaggle/input/trashclassyolo/data.yaml"
```

```python
from ultralytics import YOLO

# Load model with the last checkpoint
model = YOLO("/kaggle/input/yolo-last-checkpoint-50-epochs/pytorch/default/1/last.pt")

# Resume training
model.train(
    data=dataset_yaml,
    epochs=50,
    batch=16,
    imgsz=640,
    workers=4, 
    name="trash_detection_yolo",
    device="cuda",
    augment=True,
)
```

```python
# Evaluasi
results = model.val()
print("Validation Results:", results)

# Ekspor Model (Opsional)
model.export(format="onnx")  # Ekspor ke format ONNX
```
