```python
from ultralytics import YOLO

if __name__ == "__main__":
    # Path ke file dataset YAML
    dataset_yaml = "/content/Trash-Yolo-Subset/data.yml"

    # Load Pre-trained YOLOv8 Model
    model = YOLO("yolo11s.pt")  # YOLOv8 small pre-trained model

    # Training
    model.train(
        data=dataset_yaml,        # Path ke file YAML dataset
        epochs=50,                # Jumlah epoch
        batch=16,                  # Batch size
        imgsz=640,                # Resolusi gambar
        workers=4,                # Jumlah worker
        name="trash_detection_yolo",  # Nama eksperimen
        device="cuda",             # Gunakan CPU (ubah ke "cuda" untuk GPU)
        augment=True
    )

    # Evaluasi
    results = model.val()  # Validasi otomatis menggunakan data di file YAML
    print("Validation Results:", results)

    # Ekspor Model (Opsional)
    model.export(format="onnx")  # Ekspor ke format ONNX
```

```python
import zipfile
import os

# Nama file ZIP yang akan dibuat
zip_filename = "Trash-yolo-result-final.zip"

# Folder yang akan di-zip
folder_to_zip = "/content/runs"

# Membuat file ZIP
with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
    for root, dirs, files in os.walk(folder_to_zip):
        for file in files:
            file_path = os.path.join(root, file)
            # Menambahkan file ke dalam ZIP dengan path relatif
            zipf.write(file_path, os.path.relpath(file_path, folder_to_zip))

print(f"File ZIP '{zip_filename}' berhasil dibuat.")

# Mengunduh file ZIP
from google.colab import files
files.download(zip_filename)
```
